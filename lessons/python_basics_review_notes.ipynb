1. Variables and why they matter

They act as placeholders for data. They store values that can change, like name, type, and value

Types are flexible in Python -> you can creat custom types, and convert between most types.

    -Bools
    -Floats
    -Ints 
    -Strings

Booleans can only be True or False -> stores results of some comparison operation.

Ints and Floats are numerical data. Floats are decimaals, Intigers are whole numbers.

-Strings:

    -Text data, anything between " " or ' '
    - Used for names, messagesm text, etc + can contain text representations of numbers or boolean values

Var Examples:

    working w/:

        -booleans(bools)
        -Integers(int)
        -Floats(float)
        -Strings(str)

Python anutmatically assigns types to a value w/ simple declarations:

    name = value
    ex:

        num = 16
        james = 4
        sixteen = "farm"

bool vars work with a simple True/False setup:



```python
is_game_over = False
print(is_game_over)
is_game_over = True
print(is_game_over)

is_game_over = 5 > 6
# ^^ works w/ simple equations as well
print(is_game_over)
```
