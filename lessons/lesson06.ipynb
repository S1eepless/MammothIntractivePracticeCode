Is Linked List Cycle

lesson 06.1
-Must determine if a linked list has a cycle
if it does, return true
if it doesn't, return false







```python
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next
    
def has_cycle(head):

    slow, fast = head, head

    while fast is not None and fast.next is not None:   #if there is an end to the list the while loop will just stop

        fast = fast.next.next   #checks ahead of the next node
        slow = slow.next
        if slow == fast:        #is the slow pointer catches up to the faster pointer if there is a loop
            return True

    return False

head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = head.next.next #this line transforms the linked list into a cycle

has_cycle(head)
print(has_cycle(head))
```

Lesson 06.2

determain the length of the Linked List cycle

Provided Linked List:

Input: 4 -> -1 -> 88 -> 7 -> -2

*there is a cycle where the last value goes back to the 2nd to last value*

Output: 2

*because the cycle is just between the last two values the cycle's LENGTH is 2*

```python
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next

def calculate_cycle_length(slow):
    current = slow
    cycle_length = 0

    while True:
        current = current.next
        cycle_length += 1
        if current == slow:
            break
        
    return cycle_length


def find_length(head):

    slow, fast = head, head # both start at the head of the list

    while fast is not None and fast.next is not None:

        fast = fast.next.next
        slow = slow.next

        if slow == fast:
            return calculate_cycle_length(slow)
    return 0
        
```

```python
head = Node(4)
head.next = Node(-1)
head.next.next = Node(-88)
print(find_length(head)) # should return 0
head.next.next.next = head.next # the more '.next' after '= head.next makes the cycle shorter
print(find_length(head)) # should return 2 
```

The begining of the cycle is marked when the fast and slow pointer meet, then the program just needs to cycle using the 'current' value. Once the current value meets the slow value again, it marks the end of the cycle and adds up the amount of values it passed through on the way back to the 'slow' pointer

lesson 06.3 Find Distinct Subsets

Given a set with distinct elements
    -must find all the distinct subsets

Input: [2, 4, 3]

Output: [[], [2], [4], [2,4], [3], [2, 3], [4, 3], [2, 4, 3]]


```python
def  get_subsets(nums):
    subsets = set()

    def backtrack(index, current_subset):
        if index == len(nums):
            subsets.add(tuple(sorted(current_subset)))
            return
        
        backtrack(index + 1, current_subset + [nums[index]])
        backtrack(index + 1, current_subset)

    backtrack(0, [])
    return [list(subset) for subset in subsets]

#^^ my attempt^^

#bloated and unnecessarily complicated + not organized very well for an output

```

```python
def find_subsets(arr):

    subsets = []
    subsets.append([])

    for value in arr:

        for i in range(len(subsets)):
            set = subsets[i].copy()
            set.append(value)
            subsets.append(set)

    return subsets

# ^^provided solution^^
```
